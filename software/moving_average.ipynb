```python
#print(accuracy_percentage)
window_size = 20

def moving_avg(accuracy_percentage, window_size): # moving average function with data and window size as variables
        result = [] # empty list for the result
        moving_sum = sum(accuracy_percentage[:window_size]) # from the first window and begining of it (:) sum the values
        result.append(moving_sum / window_size) # first moving average value in the result is added
        for i in range(len(accuracy_percentage) - window_size): # loop for other windows '- window_size' substracts the first window
            moving_sum += (accuracy_percentage[i + window_size] - accuracy_percentage[i]) # adding one position to ma and substracting the first one
            result.append(moving_sum / window_size) # add the value to result variable
        return result
    
print(moving_avg(accuracy_percentage, window_size)) # prints the result

accu_ma = moving_avg(accuracy_percentage, window_size)
df = pd.DataFrame(accu_ma, columns = ['Accuracy MA'])
df.to_excel("accuracy_ma.xlsx")
```
